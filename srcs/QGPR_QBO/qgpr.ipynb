## 古典的ガウス過程回帰

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    # http://github.com/Ma-sa-ue/practice/blob/master/machine%20learning(python)/beyeisian_optimization.ipynb
    return 40.0 * np.sin(x/1.0) - np.power(0.3*(x+6.0), 2) - np.power(0.2*(x-4.0), 2) - 1.0 * np.abs(x+2.0)
```

```python
x_max = 20
x_min = -20

x_true = np.linspace(x_min, x_max, 1000)
x_true = x_true.reshape((len(x_true), 1))
y_true = f(x_true)

n_points = 20
x_sample = np.sort((x_max - x_min) * np.random.random(size=n_points) + x_min)
y_sample = f(x_sample) + np.random.normal(0, 1, size=x_sample.shape)

plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.plot(x_true, y_true, color='black', linestyle='dotted', label='True value')
plt.plot(x_sample, y_sample, 'x', color='red', label='sample point')
plt.legend()
plt.grid()
```
